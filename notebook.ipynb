menangani inaccurate value pada kolom payment_sequential (nilai maksimumnya), payment_installments (nilai minimum dan maksimumnya), dan payment_value (nilai minimum dan maksimumnya).

```python
order_payments_df[order_payments_df.payment_sequential == order_payments_df.payment_sequential.max()]
```

Nilai maksimum 29 terlihat tidak umum, karena satu pesanan biasanya memiliki lebih sedikit transaksi pembayaran berurutan. Pesanan dengan payment_sequential terlalu tinggi mungkin merupakan duplikasi atau kesalahan input.

Nilai minimum 0 menimbulkan pertanyaan, ini bisa berarti pembayaran tunai atau kesalahan input. Nilai maksimum 24 cukup tinggi, meskipun mungkin masuk akal untuk kartu kredit tertentu.

```python
order_payments_df[order_payments_df.payment_value == order_payments_df. payment_value.min()]
```

```python
order_payments_df[order_payments_df.payment_value == order_payments_df. payment_value.max()]
```

Nilai 0 berpotensi sebagai kesalahan, kecuali itu adalah diskon penuh atau voucher. Nilai maksimum 13.664,08 bisa masuk akal untuk pesanan dalam jumlah besar, tetapi perlu dicek apakah jumlah ini sesuai dengan harga pesanan terkait.

**Menggunakan metode IQR (Interquartile Range) untuk menangani inaccurate value**

```python
# Menentukan batas outlier menggunakan metode IQR
Q1 = order_payments_df[['payment_sequential', 'payment_installments', 'payment_value']].quantile(0.25)
Q3 = order_payments_df[['payment_sequential', 'payment_installments', 'payment_value']].quantile(0.75)
IQR = Q3 - Q1

# Menentukan batas bawah dan atas untuk deteksi outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menyaring data agar hanya menyertakan nilai dalam batas wajar
cleaned_order_payments_df = order_payments_df[
    (order_payments_df['payment_sequential'] <= 10) &  # Menghapus nilai ekstrem >10
    (order_payments_df['payment_installments'] > 0) &  # Menghapus installments = 0 jika tidak valid
    (order_payments_df['payment_value'] > 0) &  # Menghapus payment_value = 0 jika tidak valid
    (order_payments_df['payment_value'] >= lower_bound['payment_value']) &
    (order_payments_df['payment_value'] <= upper_bound['payment_value'])
]

# Menampilkan jumlah data sebelum dan sesudah pembersihan
original_count = order_payments_df.shape[0]
cleaned_count = cleaned_order_payments_df.shape[0]

original_count, cleaned_count
```

-  Mengganti tipe data pada kolom order_purchase_timestamp, order_approved_at, order_delivered_carrier_date, order_delivered_customer_date, dan order_estimated_delivery_date menjadi datetime.

```python
datetime_columns = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]

for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
orders_df.info()
```

- Menangani missing values pada kolom order_approved_at, order_delivered_carrier_date, dan order_delivered_customer_date.

```python
orders_df[orders_df.order_approved_at.isna()]
```

Dapat dilihat bahwa baris data tersebut masih mengandung banyak informasi penting sehingga sayang jika langsung dibuang. Oleh karena itu, pada kasus ini, kita akan menggunakan metode imputation untuk menangani missing value. Menggunakan rata-rata waktu persetujuan untuk mengisi nilai kosong.

```python
mean_approval_time = (orders_df['order_approved_at'] - orders_df['order_purchase_timestamp']).mean()
orders_df.loc[orders_df['order_approved_at'].isna(), 'order_approved_at'] = \
    orders_df['order_purchase_timestamp'] + mean_approval_time
```

```python
orders_df.isna().sum()
```

```python
orders_df[orders_df.order_delivered_carrier_date.isna()]
```

Dapat dilihat bahwa baris data tersebut masih mengandung banyak informasi penting sehingga sayang jika langsung dibuang. Oleh karena itu, pada kasus ini, kita akan menggunakan metode imputation untuk menangani missing value. Menggunakan rata-rata waktu pengiriman ke kurir untuk mengisi nilai kosong.

```python
mean_carrier_time = (orders_df['order_delivered_carrier_date'] - orders_df['order_approved_at']).mean()
orders_df.loc[orders_df['order_delivered_carrier_date'].isna(), 'order_delivered_carrier_date'] = \
    orders_df['order_approved_at'] + mean_carrier_time
```

```python
orders_df.isna().sum()
```

```python
orders_df[orders_df.order_delivered_customer_date.isna()]
```

Dapat dilihat bahwa baris data tersebut masih mengandung banyak informasi penting sehingga sayang jika langsung dibuang. Oleh karena itu, pada kasus ini, kita akan menggunakan metode imputation untuk menangani missing value. Menggunakan rata-rata waktu pengiriman ke pelanggan untuk mengisi nilai kosong.

```python
mean_delivery_time = (orders_df['order_delivered_customer_date'] - orders_df['order_delivered_carrier_date']).mean()
orders_df.loc[orders_df['order_delivered_customer_date'].isna(), 'order_delivered_customer_date'] = \
    orders_df['order_delivered_carrier_date'] + mean_delivery_time
```

```python
orders_df.isna().sum()
```

```python
most_common_category = products_df['product_category_name'].mode()[0]
products_df['product_category_name'].fillna(most_common_category, inplace=True)
```

```python
products_df.isna().sum()
```

```python
products_df[products_df.product_name_lenght.isna()]
```

```python
products_df[products_df.product_description_lenght.isna()]
```

missing values pada kolom product_name_lenght dan product_description_lenght ditangani menggunakan metode imputation dengan cara mengisi nilai kosongnya dengan rata-rata dan pada kolom product_photos_qty missing values ditangani dengan cara mengisi nilai kosongnya dengan median.

```python
products_df['product_name_lenght'].fillna(products_df.groupby('product_category_name')['product_name_lenght'].transform('mean'), inplace=True)
products_df['product_description_lenght'].fillna(products_df.groupby('product_category_name')['product_description_lenght'].transform('mean'), inplace=True)
products_df['product_photos_qty'].fillna(products_df.groupby('product_category_name')['product_photos_qty'].transform('median'), inplace=True)
```

```python
products_df.isna().sum()
```

```python
products_df[products_df.product_weight_g.isna()]
```

```python
products_df[products_df.product_length_cm.isna()]
```

```python
products_df[products_df.product_height_cm.isna()]
```

```python
products_df[products_df.product_width_cm.isna()]
```

```python
products_df['product_weight_g'].fillna(products_df['product_weight_g'].median(), inplace=True)
products_df['product_length_cm'].fillna(products_df['product_length_cm'].median(), inplace=True)
products_df['product_height_cm'].fillna(products_df['product_height_cm'].median(), inplace=True)
products_df['product_width_cm'].fillna(products_df['product_width_cm'].median(), inplace=True)
```

## RFM Analysis

Saya akan melakukan RFM Analysis dengan langkah-langkah berikut:

1. Menghitung Recency: Jumlah hari sejak transaksi terakhir pelanggan.
2. Menghitung Frequency: Jumlah total transaksi pelanggan.
3. Menghitung Monetary: Total uang yang dihabiskan oleh pelanggan.

```python
import datetime as dt

# Tentukan tanggal referensi (misalnya, hari terakhir dalam dataset)
reference_date = orders_df["order_purchase_timestamp"].max()

# Hitung Recency (berapa hari sejak pembelian terakhir)
recency_df = orders_df.groupby("customer_id")["order_purchase_timestamp"].max().reset_index()
recency_df["recency"] = (reference_date - recency_df["order_purchase_timestamp"]).dt.days

# Hitung Frequency (jumlah total transaksi per pelanggan)
frequency_df = orders_df.groupby("customer_id")["order_id"].nunique().reset_index()
frequency_df.rename(columns={"order_id": "frequency"}, inplace=True)

# Gabungkan payments_df dengan orders_df berdasarkan order_id untuk mendapatkan customer_id
payments_merged_df = order_payments_df.merge(orders_df[["order_id", "customer_id"]], on="order_id", how="left")

# Hitung Monetary: total uang yang dihabiskan pelanggan setelah penggabungan
monetary_df = payments_merged_df.groupby("customer_id")["payment_value"].sum().reset_index()
monetary_df.rename(columns={"payment_value": "monetary"}, inplace=True)

# Gabungkan ulang dengan recency dan frequency
rfm_df = recency_df.merge(frequency_df, on="customer_id").merge(monetary_df, on="customer_id")

rfm_df.head()
```

```python
# Set gaya visualisasi
sns.set_style("whitegrid")

# Buat figure dan axis
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Histogram Recency
sns.histplot(rfm_df["recency"], bins=30, kde=True, ax=axes[0, 0], color="blue")
axes[0, 0].set_title("Distribusi Recency")
axes[0, 0].set_xlabel("Hari sejak pembelian terakhir")

# Histogram Frequency
sns.histplot(rfm_df["frequency"], bins=30, kde=True, ax=axes[0, 1], color="green")
axes[0, 1].set_title("Distribusi Frequency")
axes[0, 1].set_xlabel("Jumlah transaksi")

# Histogram Monetary
sns.histplot(rfm_df["monetary"], bins=30, kde=True, ax=axes[1, 0], color="red")
axes[1, 0].set_title("Distribusi Monetary")
axes[1, 0].set_xlabel("Total uang yang dihabiskan")

# Scatter plot Frequency vs Monetary
sns.scatterplot(x=rfm_df["frequency"], y=rfm_df["monetary"], ax=axes[1, 1], alpha=0.5, color="purple")
axes[1, 1].set_title("Hubungan Frequency dan Monetary")
axes[1, 1].set_xlabel("Frequency")
axes[1, 1].set_ylabel("Monetary")

# Sesuaikan layout
plt.tight_layout()

# Tampilkan plot
plt.show()
```

1. Distribusi Recency (Hari sejak pembelian terakhir - Grafik kiri atas)

    - Mayoritas pelanggan terakhir kali melakukan pembelian dalam rentang 0 - 400 hari.
    - Ada puncak yang cukup tinggi di sekitar 100, 200, dan 300 hari, menunjukkan ada pola tertentu dalam pembelian.
    - Semakin ke kanan, jumlah pelanggan yang masih aktif semakin sedikit.

2. Distribusi Frequency (Jumlah transaksi - Grafik kanan atas)

    - Grafik ini tampak hanya memiliki satu nilai, kemungkinan besar menunjukkan bahwa mayoritas pelanggan hanya melakukan satu kali transaksi.
    - Tidak ada variasi dalam jumlah transaksi, sehingga histogram hanya membentuk satu garis lurus. Ini menunjukkan bahwa kebanyakan pelanggan adalah pembeli yang membeli sekali saja (one-time buyers).

3. Distribusi Monetary (Total uang yang dihabiskan - Grafik kiri bawah)

    - Sebagian besar pelanggan menghabiskan uang dalam jumlah kecil, yang terlihat dari puncak tajam di bagian kiri.
    - Ada beberapa pelanggan yang menghabiskan jumlah besar, tetapi jumlah mereka sangat sedikit.
    - Distribusi ini sangat skewed ke kanan, menunjukkan adanya outlier pelanggan dengan belanja tinggi.

4. Hubungan Frequency vs Monetary (Grafik kanan bawah)

    - Grafik scatter plot ini menunjukkan hubungan antara jumlah transaksi (frequency) dan total uang yang dihabiskan (monetary).
    - Ini mengindikasikan bahwa sebagian besar pendapatan berasal dari pelanggan yang hanya melakukan satu kali pembelian, sehingga strategi untuk meningkatkan retensi pelanggan perlu diperhatikan.

**Insight** :
- Sebagian besar pelanggan hanya membeli sekali dan tidak kembali.
- Ada beberapa pelanggan dengan nilai transaksi tinggi, tetapi jumlahnya sedikit.
- Pelanggan cenderung tidak melakukan pembelian berulang dalam waktu dekat setelah transaksi pertama mereka.
- Untuk meningkatkan retensi pelanggan, strategi loyalty program atau promosi mungkin diperlukan agar mereka kembali melakukan transaksi.
